```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
import sklearn.datasets as datasets
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=datasets.load_iris()
X=pd.DataFrame(iris.data, columns=iris.feature_names)
y=iris.target
X.head()
```

```python
y
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(16,13))
sns.pairplot(X,diag_kind='kde',palette='viridis',kind='scatter',)
```

## Checking for outliers
#### From here we can infer that sepal width values have a proportion of outliers.

```python
sns.set_style('darkgrid')
cols=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
fig,ax=plt.subplots(2,2)
for i,col in enumerate(cols):
    fig.add_subplot(2,2,i+1)
    fig.set_size_inches(10,8)
    sns.boxplot(y=X[col],orient='v',
                linewidth=1.2,color='0.45',
                showmeans=True,showfliers=True)
    plt.tight_layout()
plt.tight_layout()
```

## Checking for correlation among the features

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8,6))
corr=X.corr(method='pearson')
sns.heatmap(corr,robust=True,cmap='Blues',xticklabels=True)

plt.tight_layout()
```

### Checking kernel density estimation of the features with relatively high correlation.In this case we have sepal length and petal width
###    which highly correlate with petal length

```python
sns.set_style=('darkgrid')
sns.jointplot(x='sepal length (cm)',
              y='petal length (cm)',
              data=X,kind='kde',
              height=7,
              cbar=True)
sns.jointplot(x='petal width (cm)',
              y='petal length (cm)',
              data=X,kind='kde',
              height=7,
              cbar=True)
```

## Train the model

```python
dt=DecisionTreeClassifier()
dt.fit(X,y)
print("Decision Tree classifier created")
```
